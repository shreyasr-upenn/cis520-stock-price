## WE FIND THE MOVING AVERAGE FOR OUR LABELS

Moving average is a mathematical treatment used frequently while analyzing stock prices. It is used to predict or forecast the price of the share because it is good at separating out random variations of the data. This allows the investor to use several techniques like support-resistances or MACD (Moving Average Convergence Divergence) lines to assess the nature of the future share price.

```python
def moving_average(x, w):
   return np.convolve(x, np.ones(w), 'valid') / w
```

```python
close_aapl = aapl_dataset['Close'].to_numpy()
close_tsla = tsla_dataset['Close'].to_numpy()
close_jpm = jpm_dataset['Close'].to_numpy()

mov_avg_aapl = moving_average(aapl_dataset['Close'].to_numpy(), 100)
mov_avg_tsla = moving_average(tsla_dataset['Close'].to_numpy(), 100)
mov_avg_jpm = moving_average(jpm_dataset['Close'].to_numpy(), 100)


num_zer = np.zeros(99)


mov_avg_aapl = np.concatenate((close_aapl[:99],mov_avg_aapl),axis=0)
mov_avg_tsla = np.concatenate((close_tsla[:99],mov_avg_tsla),axis=0)
mov_avg_jpm = np.concatenate((close_jpm[:99],mov_avg_jpm),axis=0)


#aapl_dataset['Close'].to_numpy().shape # 2992
```
